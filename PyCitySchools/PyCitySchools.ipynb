```python
import pandas as pd
```

```python
schools_path = 'schools_complete.csv'
students_path = 'students_complete.csv'

schools_df = pd.read_csv(schools_path)
students_df = pd.read_csv(students_path)

students_df.head()
```

```python
schools_df.head()
```

```python
schools = schools_df.school_name.unique()
print(schools)
```

```python
for school, row in schools_df.iterrows():
    schools_df.loc[school,'$ per Student'] = row['budget'] / row['size']
schools_df
    
```

```python
# Create district summary
total_students = len(students_df)
total_schools = len(schools_df)
total_budget = schools_df.budget.sum()
avg_math = round(students_df.math_score.mean(),2)
avg_reading = round(students_df.reading_score.mean(),2)
```

```python
# Calculate % who passed math
pass_math = students_df.loc[students_df.math_score >= 70,'Student ID'].count()
pct_pass_math = round((pass_math/total_students)*100,2)
pct_pass_math
```

```python
# Calculate % who passed reading
pass_reading = students_df.loc[students_df.reading_score >= 70,'Student ID'].count()
pct_pass_read = round((pass_reading/total_students)*100,2)
pct_pass_read
```

```python
# Calculate % who passed both
pass_both = students_df.loc[(students_df.math_score >=70) & (students_df.reading_score >= 70),'Student ID'].count()
pct_pass_both = round((pass_both/total_students)*100,2)
pct_pass_both
```

```python
# Create high level snapshot of district's key metrics
district_summary = pd.DataFrame({'Total Schools':[total_schools],
                             'Total Students':[total_students],
                             'Total Budget':[total_budget],
                             'Avg. Math Score':[avg_math],
                             'Avg. Reading Score':[avg_reading],
                             '% Passing Math':[pct_pass_math],
                             '% Passing Reading':[pct_pass_read],
                             '% Passing Both':[pct_pass_both]})
district_summary
```

```python
# create a grouped school summary
groupby_school = students_df.groupby(['school_name'])

#types
sizes = groupby_school['Student ID'].count()
#totalbudg
#studentbudg
mathavgs = round(groupby_school['math_score'].mean(),2)
readavgs = round(groupby_school['reading_score'].mean(),2)
```

```python
# filter the df for the math/reading score conditions
# then use groupby to find the count of students who passed math, reading, and math AND reading

# math
pass_math_df = students_df.loc[students_df.math_score >= 70]
pass_math_grouped = pass_math_df.groupby(['school_name'],sort=False)
mathpasses = round((pass_math_grouped['Student ID'].count() / sizes) * 100,2)

# reading
pass_read_df = students_df.loc[students_df.reading_score>=70]
pass_read_grouped = pass_read_df.groupby(['school_name'],sort=False)
readpasses = round((pass_read_grouped['Student ID'].count() / sizes) * 100,2)

# math AND reading
pass_both_df = students_df.loc[(students_df.math_score >= 70) 
                               & 
                               (students_df.reading_score >=70)]
pass_both_grouped = pass_both_df.groupby(['school_name'],sort=False)
bothpasses = round((pass_both_grouped['Student ID'].count() / sizes) * 100,2)
```

```python
school_summary_df = pd.DataFrame({'Math Average':mathavgs,
                              'Reading Average':readavgs,
                              '% Pass (Math)':mathpasses,
                              '% Pass (Reading)':readpasses,
                              '% Pass Overall':bothpasses})

school_summary_df
```

```python
school_summary = pd.merge(schools_df, school_summary_df, on='school_name')
school_summ_rename = school_summary.rename(columns={'school_name':'School', 
                               'type':'Type', 
                               'size':'Total Students',
                              'budget':'Total Budget'})
school_summ_final = school_summ_rename.drop(columns={'School ID'})
```

```python
# Create a table that lists avg. Reading Score for students of each grade (9th, 10th, 11th, 12th) at each school.
# Create a pandas series for each grade. Hint: use a conditional statement.
# Group each series by school
# Combine the series into a dataframe
# Optional: give the displayed data cleaner formatting

students_indexed = students_df.set_index('school_name')

grades = students_indexed.grade.unique()

for grade in grades:
    if grade == '9th':
        ninth_math = pd.Series(students_indexed.loc[students_indexed.grade == grade]['math_score'])
        ninth_math_avg = ninth_math.groupby(level=0).mean()
    elif grade == '10th':
        tenth_math = pd.Series(students_indexed.loc[students_indexed.grade == grade]['math_score'])
        tenth_math_avg = tenth_math.groupby(level=0).mean()
    elif grade == '11th':
        eleventh_math = pd.Series(students_indexed.loc[students_indexed.grade == grade]['math_score'])
        eleventh_math_avg = eleventh_math.groupby(level=0).mean()
    elif grade == '12th':
        twelfth_math = pd.Series(students_indexed.loc[students_indexed.grade == grade]['math_score'])
        twelfth_math_avg = twelfth_math.groupby(level=0).mean()

mathframe = {'9th':ninth_math_avg, '10th':tenth_math_avg, '11th':eleventh_math_avg, '12th':twelfth_math_avg}
pd.DataFrame(mathframe)
```

```python
for grade in grades:
    if grade == '9th':
        ninth_read = pd.Series(students_indexed.loc[students_indexed.grade == grade]['reading_score'])
        ninth_read_avg = ninth_read.groupby(level=0).mean()
    elif grade == '10th':
        tenth_read = pd.Series(students_indexed.loc[students_indexed.grade == grade]['reading_score'])
        tenth_read_avg = tenth_read.groupby(level=0).mean()
    elif grade == '11th':
        eleventh_read = pd.Series(students_indexed.loc[students_indexed.grade == grade]['reading_score'])
        eleventh_read_avg = eleventh_read.groupby(level=0).mean()
    elif grade == '12th':
        twelfth_read = pd.Series(students_indexed.loc[students_indexed.grade == grade]['reading_score'])
        twelfth_read_avg = twelfth_read.groupby(level=0).mean()
        
readframe = {'9th':ninth_read_avg, '10th':tenth_read_avg, 
             '11th':eleventh_read_avg, '12th':twelfth_read_avg}
pd.DataFrame(readframe)
```

```python
# Create a table that breaks down school performances based on avg spend ranges (Per Student). 
# Use 4 reasonable bins to group school spending. 

school_summ_copy = school_summ_final.copy()

bins = [0, 585, 630, 645, 660]

groups = ['>$585', '$586-630', '$631-645', '$645-670']

school_spend_binned = pd.cut(school_summ_copy['$ per Student'], bins, labels=groups)
school_summ_copy['$ Ranges (per Student)'] = school_spend_binned
scores_spending = school_summ_copy.groupby('$ Ranges (per Student)').mean().drop(axis=1, 
                                                               columns=['Total Students', 
                                                                        'Total Budget', 
                                                                        '$ per Student'])
scores_spending
```

```python
# Create a table that breaks down school performances based on SIZE
# Use 4 reasonable bins to group school size. Include in the table each of the following:
#### Average Math Score
#### Average Reading Score
#### % Passing Math
#### % Passing Reading
#### Overall Passing Rate (Average of the above two)

bins = [0,1000, 2500, 5000]

_groups = ['Small (>1000)','Medium (1000-2500)','Large (2500-5000)']

school_size_binned = pd.cut(school_summ_copy['Total Students'], bins, labels=_groups)
school_summ_copy['School Size'] = school_size_binned
scores_size = school_summ_copy.groupby('School Size').mean().drop(axis=1,
                                                   columns=['Total Students',
                                                           'Total Budget',
                                                           '$ per Student'])
scores_size
```
